```
#Import stuff
!pip install soundfile
import soundfile as sf
from random import randint
import numpy as np
import random
import itertools
import librosa
import IPython.display as ipd
import urllib
import matplotlib.pyplot as plt
from google.colab import drive
%matplotlib inline



def plot_time_series(data):
    fig = plt.figure(figsize=(14, 8))
    plt.title('Raw wave ')
    plt.ylabel('Amplitude')
    plt.plot(np.linspace(0, 1, len(data)), data)
    plt.show()
    
#Hear it ! 

  
def white_noise(sig,sr):
  wn = np.random.randn(len(sig))
  sig_wn = sig + 0.004*wn
  return  (sig_wn)

def stretch(sig,sr, rate=1): 
    sig_str = librosa.effects.time_stretch(sig, rate)
    
    return sig_str
  
def chg_pitch(sig,sr,n_steps):
  sig_pitch = librosa.effects.pitch_shift(sig, sr, n_steps=n_steps)
  return(sig_pitch)  
  
def augment(sig,sr):
  sig0=sig
  sig1=white_noise(sig,sr)
  r1=randint(7,9)/10
  r2=randint(11,13)/10
  
  sig2=stretch(sig,sr,r1)
  sig3=stretch(sig,sr,r2)
  
  p1=randint(-2,-1)
  p2=randint(1,2)
  
  sig4=chg_pitch(sig,sr,p1)
  sig5=chg_pitch(sig,sr,p2)
  
  signals=[sig0,sig1,sig2,sig3,sig4,sig5]
  
  return(signals,sr)





  
  
  
```

```
s,sr=sf.read('marvin.wav')
signals=augment(s,sr)
signals=signals[0]

ipd.Audio(signals[0],rate=sr)
ipd.Audio(signals[1],rate=sr)

  
```
